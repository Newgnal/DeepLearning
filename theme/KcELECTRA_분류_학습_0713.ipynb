```python
import pandas as pd, re, html

# 0) 데이터 로드
RAW = "/content/final_make_data.csv"  # 균등 데이터셋 사용
df = pd.read_csv(RAW)

############################################
# 1) 라이트 클렌징 함수 ---------------------
############################################
def light_clean(text: str) -> str:
    text = html.unescape(str(text))
    text = re.sub(r"<[^>]+>", " ", text)
    text = re.sub(r"\[[^\]]*\]|\([^)]*\)", " ", text)
    text = re.sub(r"https?://\S+|[A-Za-z0-9_.+-]+@\S+", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# 2) 텍스트 클렌징 적용
df["clean_text"] = df["본문"].map(light_clean)

# 3) 라벨 인코딩 (13개 테마 고정)
theme_labels = [
    "반도체·AI", "2차전지·친환경 에너지", "헬스케어·바이오", "IT·인터넷 서비스",
    "금융·보험", "소비재·엔터테인먼트", "자동차·모빌리티", "방산·항공우주",
    "부동산·리츠", "채권·금리", "환율·외환", "원자재·귀금속", "기타"
]
label2id = {label: i for i, label in enumerate(theme_labels)}
df["label"] = df["keyword"].map(label2id)

# 4) 저장
df.to_csv("/content/news_bal_30000_cleaned.csv",
          index=False,
          columns=["clean_text", "label"])
```

```python
# =========================================================
# 0. 라이브러리 & 공통 설정
# ---------------------------------------------------------
import pandas as pd, numpy as np, random, torch, re, html
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from datasets import Dataset
from transformers import (
    AutoTokenizer, AutoModelForSequenceClassification,
    DataCollatorWithPadding, TrainingArguments, Trainer
)

SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)

# =========================================================
# 1. 데이터 로드
# ---------------------------------------------------------
CSV = "/content/final.csv"          # ← title / keyword 가 있는 파일
df  = pd.read_csv(CSV)

# 2. 결측 행 제거 (title·keyword 기준)
df = df.dropna(subset=["title", "keyword"])

# =========================================================
# 2. 라벨 정의 및 매핑
# ---------------------------------------------------------
theme_labels = [
    "반도체·AI", "2차전지·친환경 에너지", "헬스케어·바이오", "IT·인터넷 서비스",
    "금융·보험", "소비재·엔터테인먼트", "자동차·모빌리티", "방산·항공우주",
    "부동산·리츠", "채권·금리", "환율·외환", "원자재·귀금속", "기타"
]
label2id = {label: i for i, label in enumerate(theme_labels)}
id2label = {i: label for i, label in enumerate(theme_labels)}
NUM_LABELS = len(label2id)

# keyword(텍스트) → label(숫자) 변환
df["label"] = df["keyword"].map(label2id)

# =========================================================
# 3. 학습/검증 분할 & HuggingFace Dataset 구성
# ---------------------------------------------------------
train_x, val_x, train_y, val_y = train_test_split(
    df["title"],                # ← 본문(또는 제목+본문) 텍스트
    df["label"],
    stratify=df["label"],
    test_size=0.2,
    random_state=SEED
)

train_ds = Dataset.from_dict({"text": train_x.tolist(), "label": train_y.tolist()})
val_ds   = Dataset.from_dict({"text": val_x.tolist(),   "label": val_y.tolist()})

# =========================================================
# 4. 토크나이징
# ---------------------------------------------------------
MODEL_NAME = "monologg/koelectra-base-v3-discriminator"
tok = AutoTokenizer.from_pretrained(MODEL_NAME)

def tok_fn(batch):
    return tok(batch["text"], padding="max_length",
               truncation=True, max_length=256)

train_ds = train_ds.map(tok_fn, batched=True, remove_columns=["text"])
val_ds   = val_ds.map(tok_fn,  batched=True, remove_columns=["text"])
train_ds.set_format("torch"); val_ds.set_format("torch")

collator = DataCollatorWithPadding(tok, return_tensors="pt")

# =========================================================
# 5. 모델 정의
# ---------------------------------------------------------
model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels = NUM_LABELS,
    id2label   = id2label,
    label2id   = label2id
)

# =========================================================
# 6. Trainer 설정
# ---------------------------------------------------------
def metrics(pred):
    y_true = pred.label_ids
    y_pred = np.argmax(pred.predictions, axis=1)
    p, r, f, _ = precision_recall_fscore_support(
        y_true, y_pred, average="weighted", zero_division=0
    )
    return {
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": p, "recall": r, "f1": f
    }

args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=10,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    save_total_limit=1,
    seed=SEED,
    report_to="none",
)

trainer = Trainer(
    model=model, args=args,
    train_dataset=train_ds, eval_dataset=val_ds,
    tokenizer=tok, data_collator=collator,
    compute_metrics=metrics
)

# =========================================================
# 7. 학습 & 평가
# ---------------------------------------------------------
trainer.train()
print("📊 Validation:", trainer.evaluate())

# =========================================================
# 8. 추론 예시
# ---------------------------------------------------------
examples = [
    "미국 국채 금리가 상승하면서 환율 변동성이 커지고 있다.",
    "AI 반도체 산업 투자 확대 발표",
    "친환경 에너지 관련 정책이 새롭게 발표되었다.",
    "KB금융은 올해 디지털 전환에 약 2,000억을 투자한다.",
    "메타버스 플랫폼이 소비재 분야에 본격 적용된다."
]

inputs = tok(examples, padding=True, truncation=True,
             return_tensors="pt", max_length=256).to(model.device)

with torch.no_grad():
    pred_ids = torch.argmax(model(**inputs).logits, dim=1).tolist()

print("예측 결과:")
for text, label_id in zip(examples, pred_ids):
    print(f"📰 '{text}'\n→ 예측 테마: {id2label[label_id]}\n")
```
